```python
# 4.Reverse a string (without slicing)

s = input("Enter string: ")
rev = ""
for char in s:
    rev = char + rev
print("Reverse:", rev)
```

```python
# 6.Check if a string is palindrome

s = input("Enter a string: ")
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# 7.Check if a string is a palindrome

s = input("Enter string: ")
rev = ""
for char in s:
    rev = char + rev
if s == rev:
    print("Palindrome")
else:
    print("Not palindrome")
```
