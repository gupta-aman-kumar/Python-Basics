```python
# 1. Print numbers 1 to 10

for i in range(1,11):
    print(i, end=" ")
```

```python
# 2.  Print numbers 10 to 1 (reverse)

for i in range(10,0,-1):
    print(i,end=" ")
```

```python
# 3. Print even numbers from 1 to 20

for i in range(1,21):
  if  i % 2 == 0:
    print(i,end=" ")
```

```python
# 4. Find sum of first n numbers

n = int(input("enter number"))
sum = 0
for i in range(1,n+1):
    sum = sum +i
print(sum)
```

```python
# 5. Find factorial of a number

n=int(input("enter number"))
fact = 1
for i in range(1,n+1):
    fact = fact * i
print(fact)
```

```python
# 6. Reverse a number

rev= (input("enter"))
for i in range(len(rev)-1,-1,-1):
    print(rev[i],end=" ")
```

```python
# 7. pattern using a for loop:

n = int(input("enter number"))
for i in range(1,n+1):
    print("*" * i)
```

```python
# 8. Print Fibonacci series up to n terms

n = int(input("Enter number of terms: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
