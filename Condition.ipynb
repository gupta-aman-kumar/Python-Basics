```python
1. Input a number → check even or odd

num = int(input("enter number"))
if num % 2 == 0:
    print("even")
else:
    print("odd")
```

```python
2. Input a number → check positive / negative / zero

num = int(input("enter number"))
if num > 0:
    print("positive")
elif num == 0:
    print("zero")
else:
    print("negative")
```

```python
3. Input year → check leap year

yrs = int(input("enter year"))
if yrs % 400 == 0 or yrs % 100 != 0 and yrs % 4 == 0:
    print("leap year")
else :
    print("not a leap year")
```

```python
4. Input temperature → print hot / normal / cold

temp = int(input("enter temprature"))
if temp > 40:
    print("hot")
elif temp >= 20:
    print("normal")
else:
    print("cold")
```
