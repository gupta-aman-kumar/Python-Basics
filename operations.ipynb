## What is Associativity?
When an expression has multiple operators with the same precedence, associativity decides whether evaluation happens:

- Left to Right (most operators)
- Right to Left (a few special operators)

| Operator Type  | Associativity |
| -------------- | ------------- |
| `+ - * / // %` | Left → Right  |
| Comparison     | Left → Right  |
| `and`, `or`    | Left → Right  |
| `**`           | Right → Left  |
| `=`            | Right → Left  |


### 1. Left-to-Right Associativity (Most Operators)
Operators like +, -, *, /, //, %, comparison operators, and logical operators (and, or) evaluate left to right.

```python
# Subtraction
result1 = 10 - 5 - 2
print(result1)


result2 = (10 - 5) - 2
print(result2)
```

### 2. Right-to-Left Associativity

Some operators evaluate right to left:
- Exponentiation **
- Assignment =
- Unary operators (+, -, ~)

```python
result = 2 ** 3 ** 2
print(result)
```

```python
result = 2 ** (3 ** 2)
print(result)
```
