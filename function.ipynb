```python
# 1.Create a function hello() that prints "Good Morning"

def hello():
    print("good morning")
hello()
```

```python
# 2.Create a function greet(name) that prints "Hello <name>".

def greet(name):
    print("hello",name)
greet('aman')
```

```python
# 3.Create a function add(a, b) that returns sum of two numbers.

def add(a,b):
    return a+b
result = add(5,4)
print(result)
```

```python
# 4.Create a function square(num) that returns square of a number.

def square(num):
    return num * num
result = square(6)
print(result)
```

```python
# 5.Create a function check_even(n) that prints "Even" or "Odd".

def check_even(n):
    if n % 2 == 0:
        print("even")
    else:
       print("odd")
check_even(5)
```

```python
# 6.Create a function factorial(n) using loop.

def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact * i
    return fact
result = factorial(5)
print(result)
```

```python
# 7.Write a lambda function to find the square of a number.

x = lambda a: a*a
print(x(5))
```

```python
# 8.Write a lambda function to add three numbers

add = lambda a,b,c: a+b+c
print(add(2,3,4))
```

```python
# 9.Write a lambda function to check if a number is even.

even = lambda x: x % 2 == 0
print(even(4))
print(even(7))

```
