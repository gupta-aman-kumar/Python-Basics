```python
 1. Create a list of 5 numbers → print first and last element

list = ['apple','mango','banana','cherry']
print(len(list))
```

```python
2. Take list from user → print length.

data = input("Enter numbers separated by space: ")
nums = data.split()     # convert to list
print(len(nums))
```

```python
3. check type

list = ["apple","mango","banana","cheery"]
print(type(list))
```

```python
4.Add an element at the end of a list

nums = [1,2,3,4]
nums.append(5)
print(nums)
```

```python
5.Print all elements of a list using a loop

nums = [1,2,3,4,5]
for n in nums:
    print(n,end=" ")
```

```python
6.Print all elements of a list using a loop

num = [1,2,3,4,5]
sum = 0
for n in num:
    sum = sum + n
print(sum)
```

```python
7.Find maximum number in list

nums = [4, 9, 2, 15, 6]
print(max(nums))
```

```python
8.Remove an element from a list

my_list = [1,2,3,4,5]
my_list.remove(4)
print(my_list)
```

```python
9.Check if a number exists in a list

nums = [1,2,3,4,10,66,8]
num = 3
if num in nums:
    print("number found")
else:
    print("number not found")
```

```python
10.to change the value of a specific item, refer to the index number:

list = [1,2,3,7,5]
list[3] = 4
print(list)

```
