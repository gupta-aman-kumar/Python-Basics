```python
# Create the set
colors = {"red", "green", "blue"}
print(colors)
```

```python
# Add "yellow"
colors.add("yellow")
print(colors)
```

```python
colors.remove("green")
print(colors)
```

```python
# Print the number of items
print(len(colors))
```

```python
# create set from list
numbers = [1,2,3,4,5,5]
numbers = set(numbers)
print(numbers)
```
