```python
# Create the dictionary

car = {
    "brand": "Toyota",
    "model": "Fortuner"
}
```

```python
# Print the model

print(car["model"])
```

```python
# Add a color key

car["color"] = "Black"
```

```python
# Remove the brand key

car.pop("brand")
```

```python
# Print the dictionary

print(car)
```
